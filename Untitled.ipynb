```python
@dataclass
class FullCellInfo:
    """Structure complÃ¨te d'une cellule Excel avec TOUTES ses informations"""
    # Contenu
    raw_value: Any = ""
    cell_type: int = 0  # 0=empty, 1=text, 2=number, 3=formula
    formula: str = ""
    
    # Position
    row: int = 0
    col: int = 0
    sheet_id: str = "default"
    
    # Style complet
    style_id: str = "s0"
    bold: bool = False
    italic: bool = False
    underline: bool = False
    font_size: int = 11
    font_family: str = "Arial"
    
    # Couleurs
    text_color: str = "#000000"
    background_color: str = "#FFFFFF" 
    
    # Bordures
    border_top: int = 0
    border_bottom: int = 0
    border_left: int = 0
    border_right: int = 0
    border_color: str = "#000000"
    
    # Alignement
    horizontal_align: int = 0  # 0=general, 1=left, 2=center, 3=right
    vertical_align: int = 0    # 0=bottom, 1=middle, 2=top
    text_wrap: bool = False
    
    # Format et mÃ©tadonnÃ©es
    number_format: str = "General"
    is_merged: bool = False
    merge_range: Tuple[int, int, int, int] = (0, 0, 0, 0)
    has_comment: bool = False
    is_locked: bool = False
    is_hidden: bool = False
    
    def __post_init__(self):
        """Post-processing aprÃ¨s crÃ©ation"""
        if isinstance(self.raw_value, (int, float)):
            self.cell_type = 2
        elif isinstance(self.raw_value, str) and self.raw_value.strip():
            self.cell_type = 1
        elif self.formula:
            self.cell_type = 3
        else:
            self.cell_type = 0
```

```python
class ExcelParser:
    """Parse les fichiers Excel JSON en structures FullCellInfo"""
    
    @staticmethod
    def parse_excel_json(excel_data: Dict) -> List[FullCellInfo]:
        """Convertit JSON Excel en liste de FullCellInfo"""
        cells = []
        
        styles = excel_data.get('styles', {})
        
        for sheet_id, sheet_info in excel_data.get('sheets', {}).items():
            cell_data = sheet_info.get('cellData', {})
            
            for row_str, row_data in cell_data.items():
                row = int(row_str)
                for col_str, cell_info in row_data.items():
                    col = int(col_str)
                    
                    # Extraire les informations de style
                    style_id = cell_info.get('s', 's0')
                    style = styles.get(style_id, {})
                    
                    cell = FullCellInfo(
                        raw_value=cell_info.get('v', ''),
                        formula=cell_info.get('f', ''),
                        row=row,
                        col=col,
                        sheet_id=sheet_id,
                        style_id=style_id,
                        **ExcelParser._parse_style(style)
                    )
                    
                    cells.append(cell)
        
        return cells
    
    @staticmethod
    def _parse_style(style: Dict) -> Dict:
        """Parse les informations de style"""
        parsed = {
            'bold': bool(style.get('bl', 0)),
            'italic': bool(style.get('it', 0)),
            'underline': bool(style.get('ul', 0)),
            'font_size': style.get('fs', 11),
            'text_wrap': bool(style.get('tb', 0)),
        }
        
        # Couleurs
        if 'cl' in style:
            parsed['text_color'] = style['cl'].get('rgb', '#000000')
        if 'bg' in style:
            parsed['background_color'] = style['bg'].get('rgb', '#FFFFFF')
            
        # Bordures
        borders = style.get('bd', {})
        parsed.update({
            'border_top': borders.get('t', {}).get('s', 0),
            'border_bottom': borders.get('b', {}).get('s', 0),
            'border_left': borders.get('l', {}).get('s', 0),
            'border_right': borders.get('r', {}).get('s', 0),
        })
        
        # Alignement
        parsed.update({
            'horizontal_align': style.get('ht', 0),
            'vertical_align': style.get('vt', 0),
        })
        
        return parsed
```

```python
%matplotlib widget

import torch
import math
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# GÃ©nÃ©rer l'encodage positionnel
def positional_encoding(d_model=16, max_len=100):
    pe = torch.zeros(max_len, d_model)
    position = torch.arange(0, max_len).unsqueeze(1).float()
    div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
    pe[:, 0::2] = torch.sin(position * div_term)
    pe[:, 1::2] = torch.cos(position * div_term)
    return pe

pe = positional_encoding()

fig, ax = plt.subplots()
lines = [ax.plot([], [], label=f"dim {i}")[0] for i in range(4)]
ax.set_xlim(0, 100)
ax.set_ylim(-1.2, 1.2)
ax.legend()
ax.set_title("Encodage positionnel sinusoÃ¯dal")
ax.set_xlabel("Position")
ax.set_ylabel("Valeur")

def init():
    for line in lines:
        line.set_data([], [])
    return lines

def animate(i):
    for j, line in enumerate(lines):
        dim = (i + j) % pe.shape[1]
        x = list(range(pe.shape[0]))
        y = pe[:, dim].numpy()
        line.set_data(x, y)
    return lines

ani = animation.FuncAnimation(fig, animate, init_func=init, frames=64,
                              interval=200, blit=True)

plt.show()
```

```python
class PositionalEncoding(nn.Module):
    """Encodage positionnel sinusoÃ¯dal pour Transformer"""
    
    def __init__(self, d_model: int, max_len: int = 5000):
        super().__init__()
        
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * 
                           (-math.log(10000.0) / d_model))
        
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        
        self.register_buffer('pe', pe)
    
    def forward(self, x):
        """x shape: (seq_len, batch_size, d_model) ou (batch_size, seq_len, d_model)"""
        if x.dim() == 3 and x.size(1) == x.size(0):  # (batch, seq, d_model)
            seq_len = x.size(1)
            return x + self.pe[:seq_len, :].unsqueeze(0)
        else:  # (seq, batch, d_model)
            seq_len = x.size(0)
            return x + self.pe[:seq_len, :].unsqueeze(1)
```

```python
class TransformerCellEmbedder(nn.Module):
    """
    Transformer multi-modal pour encoder les cellules Excel complÃ¨tes.
    Chaque cellule â†’ embedding 128D avec interactions cross-modales riches.
    """
    
    def __init__(self, 
                 embedding_dim: int = 128,
                 vocab_size: int = 10000,
                 num_heads: int = 8,
                 num_layers: int = 4,
                 max_text_length: int = 20,
                 dropout: float = 0.1):
        super().__init__()
        
        self.embedding_dim = embedding_dim
        self.vocab_size = vocab_size
        self.max_text_length = max_text_length
        self.num_heads = num_heads
        self.num_layers = num_layers
        
        # === VOCABULAIRE ET MAPPINGS ===
        self.string_to_id = {'<PAD>': 0, '<UNK>': 1, '<CLS>': 2, '<SEP>': 3}
        self.id_to_string = {v: k for k, v in self.string_to_id.items()}
        self.color_vocab = {}
        self.format_vocab = {}
        
        # === EMBEDDINGS POUR CHAQUE MODALITÃ‰ ===
        
        # 1. Texte
        self.text_embedding = nn.Embedding(vocab_size, embedding_dim)
        self.text_pos_encoding = PositionalEncoding(embedding_dim, max_text_length)
        
        # 2. Style (20 features â†’ embedding_dim)
        self.style_projector = nn.Sequential(
            nn.Linear(20, embedding_dim),
            nn.LayerNorm(embedding_dim),
            nn.GELU(),
            nn.Dropout(dropout)
        )
        
        # 3. Position spatiale
        self.spatial_row_emb = nn.Embedding(2000, embedding_dim // 2)  # Support jusqu'Ã  2000 lignes
        self.spatial_col_emb = nn.Embedding(100, embedding_dim // 2)   # Support jusqu'Ã  100 colonnes
        self.spatial_projector = nn.Linear(embedding_dim, embedding_dim)
        
        # 4. Type de cellule
        self.type_embedding = nn.Embedding(4, embedding_dim)
        
        # 5. Valeurs numÃ©riques
        self.numeric_projector = nn.Sequential(
            nn.Linear(6, embedding_dim),  # 6 features numÃ©riques
            nn.LayerNorm(embedding_dim),
            nn.Tanh()
        )
        
        # === TOKENS DE MODALITÃ‰ ===
        # Tokens apprenables pour identifier chaque type d'information
        self.modality_tokens = nn.Parameter(torch.randn(6, embedding_dim) * 0.02)
        # 0: CLS, 1: TEXT, 2: STYLE, 3: POSITION, 4: TYPE, 5: NUMERIC
        
        # === TRANSFORMER ENCODER ===
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=embedding_dim,
            nhead=num_heads,
            dim_feedforward=embedding_dim * 4,
            dropout=dropout,
            activation='gelu',
            batch_first=True,
            norm_first=True  # Pre-norm pour stabilitÃ©
        )
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        
        # === AGRÃ‰GATION FINALE ===
        self.final_aggregator = nn.Sequential(
            nn.Linear(embedding_dim, embedding_dim),
            nn.LayerNorm(embedding_dim),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(embedding_dim, embedding_dim)
        )
        
        # === NORMALISATION DES VALEURS NUMÃ‰RIQUES ===
        self.numeric_stats = {
            'min': 0, 'max': 1, 'mean': 0, 'std': 1,
            'q25': 0, 'q75': 1, 'log_min': 0, 'log_max': 1
        }
        
        self.is_fitted = False
        
    def build_vocabulary(self, cells: List[FullCellInfo], min_freq: int = 2):
        """Construit le vocabulaire Ã  partir des cellules"""
        print(f"ðŸ”¨ Construction du vocabulaire Ã  partir de {len(cells)} cellules...")
        
        # Compter les tokens textuels
        text_counter = Counter()
        color_set = set()
        format_set = set()
        numeric_values = []  # GARDER COMME LISTE PYTHON
        
        for cell in cells:
            # Texte
            if cell.cell_type == 1 and cell.raw_value:
                words = str(cell.raw_value).strip().split()
                text_counter.update(words)
            
            # Couleurs
            color_set.add(cell.text_color)
            color_set.add(cell.background_color)
            color_set.add(cell.border_color)
            
            # Formats
            format_set.add(cell.number_format)
            format_set.add(cell.font_family)
            
            # Valeurs numÃ©riques
            if cell.cell_type == 2:
                try:
                    val = float(cell.raw_value)
                    if not (np.isnan(val) or np.isinf(val)):
                        numeric_values.append(val)  # AJOUTER Ã€ LA LISTE
                except:
                    pass
        
        # Construire vocabulaire textuel
        current_id = len(self.string_to_id)
        for word, freq in text_counter.items():
            if freq >= min_freq and current_id < self.vocab_size:
                self.string_to_id[word] = current_id
                self.id_to_string[current_id] = word
                current_id += 1
        
        # Vocabulaires de couleurs et formats (hashing)
        self.color_vocab = {color: self._hash_string(color) for color in color_set}
        self.format_vocab = {fmt: self._hash_string(fmt) for fmt in format_set}
        
        # Statistiques numÃ©riques - FIX ICI
        if len(numeric_values) > 0:  # âœ… CORRECTION: Utiliser len() au lieu de if numeric_values:
            numeric_array = np.array(numeric_values)  # âœ… Convertir en array SEULEMENT si nÃ©cessaire
            self.numeric_stats = {
                'min': float(np.min(numeric_array)),
                'max': float(np.max(numeric_array)),
                'mean': float(np.mean(numeric_array)),
                'std': float(np.std(numeric_array)),
                'q25': float(np.percentile(numeric_array, 25)),
                'q75': float(np.percentile(numeric_array, 75)),
                'log_min': float(np.log10(np.maximum(numeric_array, 1e-8)).min()),
                'log_max': float(np.log10(np.maximum(numeric_array, 1e-8)).max())
            }
        
        self.is_fitted = True
        
        print(f"âœ… Vocabulaire construit:")
        print(f"   - Mots textuels: {len(self.string_to_id)} (coverage: {current_id}/{len(text_counter)})")
        print(f"   - Couleurs uniques: {len(color_set)}")
        print(f"   - Formats uniques: {len(format_set)}")
        print(f"   - Valeurs numÃ©riques: {len(numeric_values)}")
        if len(numeric_values) > 0:  # âœ… CORRECTION: Utiliser len() ici aussi
            print(f"   - Range numÃ©rique: [{self.numeric_stats['min']:.2f}, {self.numeric_stats['max']:.2f}]")
    
    def forward(self, cells_batch: List[FullCellInfo]) -> torch.Tensor:
        """
        Forward pass principal : cellules â†’ embeddings unifiÃ©s
        
        Args:
            cells_batch: Liste de cellules avec toutes leurs informations
            
        Returns:
            Tensor (batch_size, embedding_dim) : embeddings unifiÃ©s
        """
        if not self.is_fitted:
            raise ValueError("Le modÃ¨le doit Ãªtre entraÃ®nÃ© avec build_vocabulary() d'abord")
        
        batch_size = len(cells_batch)
        device = next(self.parameters()).device
        
        # === Ã‰TAPE 1: CRÃ‰ER LES SÃ‰QUENCES MULTI-MODALES ===
        
        all_sequences = []
        all_masks = []
        
        for cell in cells_batch:
            sequence_tokens = []
            
            # 1. Token CLS (dÃ©but de sÃ©quence)
            cls_token = self.modality_tokens[0].unsqueeze(0)  # (1, embedding_dim)
            sequence_tokens.append(cls_token)
            
            # 2. Tokens textuels
            text_tokens = self._encode_text_tokens(cell)
            if text_tokens.size(0) > 0:
                text_tokens = text_tokens + self.modality_tokens[1].unsqueeze(0)
                sequence_tokens.append(text_tokens)
            
            # 3. Token style
            style_token = self._encode_style_token(cell)
            style_token = style_token + self.modality_tokens[2]
            sequence_tokens.append(style_token.unsqueeze(0))
            
            # 4. Token position
            pos_token = self._encode_position_token(cell)
            pos_token = pos_token + self.modality_tokens[3]
            sequence_tokens.append(pos_token.unsqueeze(0))
            
            # 5. Token type
            type_token = self._encode_type_token(cell)
            type_token = type_token + self.modality_tokens[4]
            sequence_tokens.append(type_token.unsqueeze(0))
            
            # 6. Token numÃ©rique
            numeric_token = self._encode_numeric_token(cell)
            numeric_token = numeric_token + self.modality_tokens[5]
            sequence_tokens.append(numeric_token.unsqueeze(0))
            
            # ConcatÃ©ner tous les tokens
            cell_sequence = torch.cat(sequence_tokens, dim=0)  # (seq_len, embedding_dim)
            all_sequences.append(cell_sequence)
        
        # === Ã‰TAPE 2: PADDING ET BATCHING ===
        
        max_seq_len = max(seq.size(0) for seq in all_sequences)
        padded_sequences = []
        attention_masks = []
        
        for seq in all_sequences:
            seq_len = seq.size(0)
            
            if seq_len < max_seq_len:
                # Padding avec des zÃ©ros
                padding = torch.zeros(max_seq_len - seq_len, self.embedding_dim, 
                                    device=device, dtype=seq.dtype)
                padded_seq = torch.cat([seq, padding], dim=0)
                
                # Masque d'attention
                mask = torch.cat([
                    torch.ones(seq_len, device=device, dtype=torch.bool),
                    torch.zeros(max_seq_len - seq_len, device=device, dtype=torch.bool)
                ])
            else:
                padded_seq = seq
                mask = torch.ones(seq_len, device=device, dtype=torch.bool)
            
            padded_sequences.append(padded_seq)
            attention_masks.append(mask)
        
        # Stack en batch
        batch_sequences = torch.stack(padded_sequences).to(device)  # (batch, seq_len, emb_dim)
        batch_masks = torch.stack(attention_masks).to(device)       # (batch, seq_len)
        
        # === Ã‰TAPE 3: TRANSFORMER MULTI-MODAL ===
        
        # Masque pour padding (True = ignorer, False = attention)
        src_key_padding_mask = ~batch_masks
        
        # Passage dans le Transformer
        transformer_output = self.transformer(
            batch_sequences,
            src_key_padding_mask=src_key_padding_mask
        )  # (batch_size, seq_len, embedding_dim)
        
        # === Ã‰TAPE 4: AGRÃ‰GATION EN EMBEDDING UNIFIÃ‰ ===
        
        # Utiliser le token CLS pour l'embedding final (comme BERT)
        cls_embeddings = transformer_output[:, 0, :]  # (batch_size, embedding_dim)
        
        # Transformation finale
        final_embeddings = self.final_aggregator(cls_embeddings)
        
        return final_embeddings
    
    def _encode_text_tokens(self, cell: FullCellInfo) -> torch.Tensor:
        """Encode les tokens textuels avec positional encoding"""
        device = next(self.parameters()).device
        
        if cell.cell_type != 1 or not cell.raw_value:
            return torch.empty(0, self.embedding_dim, device=device)
        
        # Tokenisation
        text = str(cell.raw_value).strip()
        words = text.split()[:self.max_text_length-2]  # -2 pour SEP
        
        # Conversion en IDs
        token_ids = []
        for word in words:
            token_ids.append(self.string_to_id.get(word, self.string_to_id['<UNK>']))
        token_ids.append(self.string_to_id['<SEP>'])
        
        if not token_ids:
            return torch.empty(0, self.embedding_dim, device=device)
        
        # Embeddings
        token_ids_tensor = torch.tensor(token_ids, device=device)
        embeddings = self.text_embedding(token_ids_tensor)
        
        # Positional encoding
        embeddings = self.text_pos_encoding(embeddings.unsqueeze(0)).squeeze(0)
        
        return embeddings
    
    def _encode_style_token(self, cell: FullCellInfo) -> torch.Tensor:
        """Encode les informations de style en un token"""
        device = next(self.parameters()).device
        
        features = [
            float(cell.bold),
            float(cell.italic), 
            float(cell.underline),
            cell.font_size / 24.0,  # Normalisation
            self._normalize_hash(self.color_vocab.get(cell.text_color, 0)),
            self._normalize_hash(self.color_vocab.get(cell.background_color, 0)),
            self._normalize_hash(self.color_vocab.get(cell.border_color, 0)),
            cell.border_top / 8.0,
            cell.border_bottom / 8.0,
            cell.border_left / 8.0,
            cell.border_right / 8.0,
            cell.horizontal_align / 3.0,
            cell.vertical_align / 2.0,
            float(cell.text_wrap),
            float(cell.is_merged),
            float(cell.has_comment),
            float(cell.is_locked),
            float(cell.is_hidden),
            self._normalize_hash(self.format_vocab.get(cell.number_format, 0)),
            self._normalize_hash(self.format_vocab.get(cell.font_family, 0))
        ]
        
        features_tensor = torch.tensor(features, device=device, dtype=torch.float)
        return self.style_projector(features_tensor)
    
    def _encode_position_token(self, cell: FullCellInfo) -> torch.Tensor:
        """Encode la position spatiale"""
        device = next(self.parameters()).device
        
        row_emb = self.spatial_row_emb(torch.tensor(min(cell.row, 1999), device=device))
        col_emb = self.spatial_col_emb(torch.tensor(min(cell.col, 99), device=device))
        
        pos_concat = torch.cat([row_emb, col_emb])
        return self.spatial_projector(pos_concat)
    
    def _encode_type_token(self, cell: FullCellInfo) -> torch.Tensor:
        """Encode le type de cellule"""
        device = next(self.parameters()).device
        return self.type_embedding(torch.tensor(cell.cell_type, device=device))
    
    def _encode_numeric_token(self, cell: FullCellInfo) -> torch.Tensor:
        """Encode les valeurs numÃ©riques"""
        device = next(self.parameters()).device
        
        if cell.cell_type == 2:
            try:
                value = float(cell.raw_value)
                features = self._extract_numeric_features(value)
            except:
                features = [0.0] * 6
        else:
            features = [0.0] * 6
        
        features_tensor = torch.tensor(features, device=device, dtype=torch.float)
        return self.numeric_projector(features_tensor)
    
    def _extract_numeric_features(self, value: float) -> List[float]:
        """Extrait des features d'une valeur numÃ©rique"""
        stats = self.numeric_stats
        
        features = []
        
        # 1. Normalisation min-max
        if stats['max'] > stats['min']:
            minmax_norm = (value - stats['min']) / (stats['max'] - stats['min'])
            features.append(np.clip(minmax_norm, 0, 1))
        else:
            features.append(0.5)
        
        # 2. Z-score normalisÃ©
        if stats['std'] > 0:
            zscore = (value - stats['mean']) / stats['std']
            zscore_norm = np.tanh(zscore / 3) * 0.5 + 0.5
            features.append(zscore_norm)
        else:
            features.append(0.5)
        
        # 3. Log-scale
        if value > 0 and stats['log_max'] > stats['log_min']:
            log_val = np.log10(max(value, 1e-8))
            log_norm = (log_val - stats['log_min']) / (stats['log_max'] - stats['log_min'])
            features.append(np.clip(log_norm, 0, 1))
        else:
            features.append(0.0)
        
        # 4. Signe et propriÃ©tÃ©s
        features.append(1.0 if value > 0 else 0.0)  # Positif
        features.append(1.0 if value == int(value) else 0.0)  # Entier
        features.append(min(len(str(int(abs(value)))) / 10.0, 1.0))  # Nb chiffres
        
        return features
    
    def _hash_string(self, s: str) -> int:
        """Hash stable d'une string"""
        return int(hashlib.md5(s.encode()).hexdigest()[:8], 16)
    
    def _normalize_hash(self, hash_val: int) -> float:
        """Normalise un hash en [0,1]"""
        return (hash_val % 10000) / 10000.0
```

```python
class EmbeddingGraphBuilder:
    """Construit un graphe Ã  partir d'embeddings de cellules"""
    
    def __init__(self, similarity_threshold: float = 0.5):
        self.similarity_threshold = similarity_threshold
        
    def create_embedding_graph(self, embeddings: torch.Tensor, 
                             cells: List[FullCellInfo]) -> Dict:
        """
        CrÃ©e un graphe d'embeddings avec relations automatiques
        
        Args:
            embeddings: Tensor (n_cells, embedding_dim)
            cells: Liste des cellules correspondantes
            
        Returns:
            Dict avec nodes, edges, et metadata
        """
        n_cells = embeddings.shape[0]
        
        # Calcul des similaritÃ©s
        similarities = cosine_similarity(embeddings.detach().cpu().numpy())
        
        # Construction des arÃªtes
        edges = []
        edge_weights = []
        edge_types = []
        
        for i in range(n_cells):
            for j in range(i + 1, n_cells):
                cell1, cell2 = cells[i], cells[j]
                sim_score = similarities[i, j]
                
                # Relations spatiales
                spatial_relations = self._get_spatial_relations(cell1, cell2)
                
                # Relations sÃ©mantiques (via similaritÃ© d'embeddings)
                if sim_score > self.similarity_threshold:
                    edges.append((i, j))
                    edge_weights.append(sim_score)
                    edge_types.append('semantic_similarity')
                
                # Relations spatiales
                for relation, weight in spatial_relations:
                    edges.append((i, j))
                    edge_weights.append(weight)
                    edge_types.append(relation)
        
        return {
            'nodes': embeddings,
            'edges': edges,
            'edge_weights': edge_weights,
            'edge_types': edge_types,
            'cells': cells,
            'similarities': similarities
        }
    
    def _get_spatial_relations(self, cell1: FullCellInfo, 
                             cell2: FullCellInfo) -> List[Tuple[str, float]]:
        """DÃ©termine les relations spatiales entre deux cellules"""
        relations = []
        
        row_diff = abs(cell1.row - cell2.row)
        col_diff = abs(cell1.col - cell2.col)
        
        # Adjacence directe
        if (row_diff == 1 and col_diff == 0) or (row_diff == 0 and col_diff == 1):
            relations.append(('spatial_adjacent', 1.0))
        
        # MÃªme ligne
        elif row_diff == 0 and col_diff <= 5:
            weight = 1.0 / (1 + col_diff)
            relations.append(('spatial_same_row', weight))
        
        # MÃªme colonne
        elif col_diff == 0 and row_diff <= 10:
            weight = 1.0 / (1 + row_diff)
            relations.append(('spatial_same_col', weight))
        
        return relations
```

```python
# ============================================================================
# TRAITEMENT EN LOT DE FICHIERS EXCEL JSON - IMPLÃ‰MENTATION COMPLÃˆTE
# ============================================================================
# Traite une dizaine de fichiers Excel JSON en parallÃ¨le avec analyse comparative
# ============================================================================

import os
import json
import glob
from pathlib import Path
from typing import List, Dict, Tuple, Optional
import pandas as pd
import numpy as np
import torch
from concurrent.futures import ThreadPoolExecutor, as_completed
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')

# RÃ©utiliser les classes prÃ©cÃ©dentes
# from previous_cells import FullCellInfo, ExcelParser, TransformerCellEmbedder, etc.

class BatchExcelProcessor:
    """
    Processeur en lot pour analyser plusieurs fichiers Excel JSON
    avec le TransformerCellEmbedder
    """
    
    def __init__(self, 
                 model_config: Dict = None,
                 batch_size: int = 32,
                 max_workers: int = 4):
        
        self.batch_size = batch_size
        self.max_workers = max_workers
        
        # Configuration par dÃ©faut du modÃ¨le
        self.model_config = model_config or {
            'embedding_dim': 128,
            'num_heads': 8,
            'num_layers': 4,
            'vocab_size': 10000,
            'dropout': 0.1
        }
        
        # ModÃ¨le et donnÃ©es
        self.model = None
        self.all_cells = []
        self.file_metadata = {}
        self.embeddings_by_file = {}
        self.global_embeddings = None
        
    def process_excel_files(self, 
                           file_paths: List[str],
                           output_dir: str = "excel_analysis_results") -> Dict:
        """
        Traite un lot de fichiers Excel JSON
        
        Args:
            file_paths: Liste des chemins vers les fichiers JSON
            output_dir: Dossier de sortie pour les rÃ©sultats
            
        Returns:
            Dict avec rÃ©sultats de l'analyse
        """
        
        print(f"ðŸš€ TRAITEMENT EN LOT DE {len(file_paths)} FICHIERS EXCEL")
        print("=" * 60)
        
        # CrÃ©er le dossier de sortie
        Path(output_dir).mkdir(exist_ok=True)
        
        # 1. Charger et parser tous les fichiers
        print("\nðŸ“‚ 1. Chargement des fichiers...")
        all_cells_data = self._load_excel_files(file_paths)
        
        # 2. Construire le vocabulaire global
        print("\nðŸ“š 2. Construction du vocabulaire global...")
        self._build_global_model(all_cells_data)
        
        # 3. GÃ©nÃ©rer les embeddings pour tous les fichiers
        print("\nðŸ§  3. GÃ©nÃ©ration des embeddings...")
        self._generate_all_embeddings(all_cells_data)
        
        # 4. Analyses comparatives
        print("\nðŸ“Š 4. Analyses comparatives...")
        analysis_results = self._perform_comparative_analysis()
        
        # 5. Visualisations et rapports
        print("\nðŸ“ˆ 5. GÃ©nÃ©ration des visualisations...")
        self._generate_visualizations(output_dir)
        
        # 6. Sauvegarder les rÃ©sultats
        print("\nðŸ’¾ 6. Sauvegarde des rÃ©sultats...")
        self._save_results(analysis_results, output_dir)
        
        print(f"\nâœ… Traitement terminÃ© ! RÃ©sultats dans: {output_dir}")
        
        return analysis_results
    
    def _load_excel_files(self, file_paths: List[str]) -> Dict[str, List[FullCellInfo]]:
        """Charge tous les fichiers Excel en parallÃ¨le"""
        
        all_cells_data = {}
        
        def load_single_file(file_path: str) -> Tuple[str, List[FullCellInfo], Dict]:
            """Charge un seul fichier"""
            try:
                with open(file_path, 'r', encoding='utf-8') as f:
                    excel_data = json.load(f)
                
                # Parser le fichier
                cells = ExcelParser.parse_excel_json(excel_data)
                
                # MÃ©tadonnÃ©es du fichier
                metadata = {
                    'file_name': os.path.basename(file_path),
                    'file_path': file_path,
                    'total_cells': len(cells),
                    'text_cells': sum(1 for c in cells if c.cell_type == 1),
                    'numeric_cells': sum(1 for c in cells if c.cell_type == 2),
                    'empty_cells': sum(1 for c in cells if c.cell_type == 0),
                    'formula_cells': sum(1 for c in cells if c.cell_type == 3),
                    'unique_styles': len(set(c.style_id for c in cells)),
                    'sheets': list(excel_data.get('sheets', {}).keys())
                }
                
                return file_path, cells, metadata
                
            except Exception as e:
                print(f"âŒ Erreur lors du chargement de {file_path}: {e}")
                return file_path, [], {}
        
        # Traitement en parallÃ¨le
        with ThreadPoolExecutor(max_workers=self.max_workers) as executor:
            future_to_file = {
                executor.submit(load_single_file, file_path): file_path 
                for file_path in file_paths
            }
            
            for future in tqdm(as_completed(future_to_file), 
                             total=len(file_paths), 
                             desc="Chargement fichiers"):
                file_path, cells, metadata = future.result()
                
                if cells:  # Fichier chargÃ© avec succÃ¨s
                    all_cells_data[file_path] = cells
                    self.file_metadata[file_path] = metadata
                    self.all_cells.extend(cells)
        
        print(f"   âœ… {len(all_cells_data)} fichiers chargÃ©s avec succÃ¨s")
        print(f"   âœ… {len(self.all_cells)} cellules totales extraites")
        
        return all_cells_data
    
    def _build_global_model(self, all_cells_data: Dict[str, List[FullCellInfo]]):
        """Construit le modÃ¨le avec vocabulaire global"""
        
        print("   ðŸ”¨ Initialisation du TransformerCellEmbedder...")
        self.model = TransformerCellEmbedder(**self.model_config).to(device)
        
        print("   ðŸ“– Construction du vocabulaire global...")
        self.model.build_vocabulary(self.all_cells)
        
        print(f"   âœ… ModÃ¨le prÃªt avec vocabulaire de {len(self.model.string_to_id)} termes")
    
    def _generate_all_embeddings(self, all_cells_data: Dict[str, List[FullCellInfo]]):
        """GÃ©nÃ¨re les embeddings pour tous les fichiers"""
        
        self.model.eval()
        all_embeddings = []
        
        with torch.no_grad():
            for file_path, cells in tqdm(all_cells_data.items(), desc="GÃ©nÃ©ration embeddings"):
                
                if not cells:
                    continue
                
                # Traiter par batches pour Ã©viter les problÃ¨mes de mÃ©moire
                file_embeddings = []
                
                for i in range(0, len(cells), self.batch_size):
                    batch_cells = cells[i:i + self.batch_size]
                    batch_embeddings = self.model(batch_cells)
                    file_embeddings.append(batch_embeddings)
                
                # ConcatÃ©ner les embeddings du fichier
                if file_embeddings:
                    file_embeddings_tensor = torch.cat(file_embeddings, dim=0)
                    self.embeddings_by_file[file_path] = file_embeddings_tensor
                    all_embeddings.append(file_embeddings_tensor)
        
        # ConcatÃ©ner tous les embeddings
        if all_embeddings:
            self.global_embeddings = torch.cat(all_embeddings, dim=0)
        
        print(f"   âœ… Embeddings gÃ©nÃ©rÃ©s: {self.global_embeddings.shape}")
    
    def _perform_comparative_analysis(self) -> Dict:
        """Effectue des analyses comparatives entre fichiers"""
        
        analysis_results = {
            'file_statistics': {},
            'similarity_matrix': None,
            'clusters': {},
            'vocabulary_analysis': {},
            'embedding_statistics': {}
        }
        
        # === 1. STATISTIQUES PAR FICHIER ===
        print("   ðŸ“‹ Calcul des statistiques par fichier...")
        
        for file_path, metadata in self.file_metadata.items():
            embeddings = self.embeddings_by_file.get(file_path)
            
            if embeddings is not None:
                # Statistiques des embeddings
                embedding_stats = {
                    'mean_norm': float(embeddings.norm(dim=1).mean()),
                    'std_norm': float(embeddings.norm(dim=1).std()),
                    'embedding_diversity': float(embeddings.std(dim=0).mean()),
                    'embedding_shape': list(embeddings.shape)
                }
                
                # Combiner avec mÃ©tadonnÃ©es
                analysis_results['file_statistics'][file_path] = {
                    **metadata,
                    **embedding_stats
                }
        
        # === 2. MATRICE DE SIMILARITÃ‰ ENTRE FICHIERS ===
        print("   ðŸ” Calcul de la matrice de similaritÃ©...")
        
        file_paths = list(self.embeddings_by_file.keys())
        n_files = len(file_paths)
        similarity_matrix = np.zeros((n_files, n_files))
        
        for i, file1 in enumerate(file_paths):
            for j, file2 in enumerate(file_paths):
                if i <= j:
                    emb1 = self.embeddings_by_file[file1]
                    emb2 = self.embeddings_by_file[file2]
                    
                    # SimilaritÃ© moyenne entre les embeddings des deux fichiers
                    similarity = torch.cosine_similarity(
                        emb1.mean(dim=0, keepdim=True),
                        emb2.mean(dim=0, keepdim=True)
                    ).item()
                    
                    similarity_matrix[i, j] = similarity
                    similarity_matrix[j, i] = similarity
        
        analysis_results['similarity_matrix'] = {
            'matrix': similarity_matrix,
            'file_names': [os.path.basename(f) for f in file_paths],
            'file_paths': file_paths
        }
        
        # === 3. CLUSTERING DES FICHIERS ===
        print("   ðŸŽ¯ Clustering des fichiers...")
        
        # ReprÃ©sentation de chaque fichier par la moyenne de ses embeddings
        file_representations = []
        for file_path in file_paths:
            emb = self.embeddings_by_file[file_path]
            file_rep = emb.mean(dim=0).cpu().numpy()
            file_representations.append(file_rep)
        
        file_representations = np.array(file_representations)
        
        # K-means clustering
        n_clusters = min(3, n_files)  # Au maximum 3 clusters
        if n_files > 1:
            kmeans = KMeans(n_clusters=n_clusters, random_state=42)
            cluster_labels = kmeans.fit_predict(file_representations)
            
            analysis_results['clusters'] = {
                'labels': cluster_labels.tolist(),
                'centers': kmeans.cluster_centers_.tolist(),
                'n_clusters': n_clusters,
                'file_to_cluster': {
                    os.path.basename(file_paths[i]): int(cluster_labels[i])
                    for i in range(len(file_paths))
                }
            }
        
        # === 4. ANALYSE DU VOCABULAIRE ===
        print("   ðŸ”¤ Analyse du vocabulaire...")
        
        vocab_analysis = {
            'total_vocabulary_size': len(self.model.string_to_id),
            'most_common_terms': [],
            'coverage_by_file': {}
        }
        
        # Coverage du vocabulaire par fichier
        for file_path, cells in zip(file_paths, [self.embeddings_by_file.keys()]):
            # Compter les termes uniques dans ce fichier
            file_terms = set()
            for cell in cells if hasattr(cells, '__iter__') else []:
                if hasattr(cell, 'raw_value') and cell.cell_type == 1:
                    words = str(cell.raw_value).split()
                    file_terms.update(words)
            
            coverage = len(file_terms & set(self.model.string_to_id.keys()))
            vocab_analysis['coverage_by_file'][os.path.basename(file_path)] = {
                'unique_terms': len(file_terms),
                'vocabulary_coverage': coverage,
                'coverage_ratio': coverage / len(file_terms) if file_terms else 0
            }
        
        analysis_results['vocabulary_analysis'] = vocab_analysis
        
        # === 5. STATISTIQUES GLOBALES DES EMBEDDINGS ===
        print("   ðŸ“Š Statistiques globales des embeddings...")
        
        if self.global_embeddings is not None:
            embedding_stats = {
                'global_shape': list(self.global_embeddings.shape),
                'mean_norm': float(self.global_embeddings.norm(dim=1).mean()),
                'std_norm': float(self.global_embeddings.norm(dim=1).std()),
                'dimension_variance': self.global_embeddings.var(dim=0).cpu().numpy().tolist(),
                'embedding_diversity': float(self.global_embeddings.std(dim=0).mean())
            }
            
            analysis_results['embedding_statistics'] = embedding_stats
        
        return analysis_results
    
    def _generate_visualizations(self, output_dir: str):
        """GÃ©nÃ¨re les visualisations"""
        
        plt.style.use('default')
        
        # === 1. MATRICE DE SIMILARITÃ‰ ===
        if 'similarity_matrix' in self.__dict__ and hasattr(self, 'file_metadata'):
            
            fig, ax = plt.subplots(figsize=(12, 10))
            
            # Utiliser les donnÃ©es de l'analyse
            if hasattr(self, '_last_analysis_results'):
                sim_data = self._last_analysis_results.get('similarity_matrix', {})
                similarity_matrix = sim_data.get('matrix')
                file_names = sim_data.get('file_names', [])
                
                if similarity_matrix is not None:
                    sns.heatmap(similarity_matrix, 
                              xticklabels=file_names,
                              yticklabels=file_names,
                              annot=True, 
                              cmap='viridis',
                              center=0.5,
                              ax=ax)
                    
                    ax.set_title('Matrice de SimilaritÃ© entre Fichiers Excel', fontsize=14, fontweight='bold')
                    plt.xticks(rotation=45, ha='right')
                    plt.yticks(rotation=0)
                    plt.tight_layout()
                    plt.savefig(f"{output_dir}/similarity_matrix.png", dpi=300, bbox_inches='tight')
                    plt.close()
        
        # === 2. DISTRIBUTION DES EMBEDDINGS (t-SNE) ===
        if self.global_embeddings is not None and len(self.embeddings_by_file) > 1:
            
            print("   ðŸ—ºï¸ GÃ©nÃ©ration de la visualisation t-SNE...")
            
            # Ã‰chantillonner pour t-SNE si trop de points
            embeddings_sample = self.global_embeddings.cpu().numpy()
            if embeddings_sample.shape[0] > 1000:
                indices = np.random.choice(embeddings_sample.shape[0], 1000, replace=False)
                embeddings_sample = embeddings_sample[indices]
            
            # t-SNE
            tsne = TSNE(n_components=2, random_state=42, perplexity=min(30, embeddings_sample.shape[0]-1))
            embeddings_2d = tsne.fit_transform(embeddings_sample)
            
            # Assigner les couleurs par fichier
            file_colors = []
            color_map = plt.cm.Set3(np.linspace(0, 1, len(self.embeddings_by_file)))
            
            current_idx = 0
            for i, (file_path, embeddings) in enumerate(self.embeddings_by_file.items()):
                file_size = embeddings.shape[0]
                sample_size = min(file_size, 1000 // len(self.embeddings_by_file))
                file_colors.extend([color_map[i]] * sample_size)
                current_idx += sample_size
                if current_idx >= 1000:
                    break
            
            # Plot
            fig, ax = plt.subplots(figsize=(12, 8))
            
            for i, (file_path, _) in enumerate(self.embeddings_by_file.items()):
                mask = np.array([j for j, c in enumerate(file_colors) if np.array_equal(c, color_map[i])])
                if len(mask) > 0:
                    ax.scatter(embeddings_2d[mask, 0], embeddings_2d[mask, 1], 
                             c=[color_map[i]], 
                             label=os.path.basename(file_path)[:20],
                             alpha=0.6, s=20)
            
            ax.set_title('Distribution des Embeddings de Cellules (t-SNE)', fontsize=14, fontweight='bold')
            ax.set_xlabel('Dimension 1')
            ax.set_ylabel('Dimension 2')
            ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
            plt.tight_layout()
            plt.savefig(f"{output_dir}/embeddings_tsne.png", dpi=300, bbox_inches='tight')
            plt.close()
        
        # === 3. STATISTIQUES DES FICHIERS ===
        print("   ðŸ“Š GÃ©nÃ©ration des graphiques statistiques...")
        
        # Graphique en barres des statistiques par fichier
        if self.file_metadata:
            
            file_names = [os.path.basename(f) for f in self.file_metadata.keys()]
            total_cells = [meta['total_cells'] for meta in self.file_metadata.values()]
            text_cells = [meta['text_cells'] for meta in self.file_metadata.values()]
            numeric_cells = [meta['numeric_cells'] for meta in self.file_metadata.values()]
            
            fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))
            
            # Graphique 1: Nombre total de cellules
            ax1.bar(range(len(file_names)), total_cells, color='skyblue', alpha=0.7)
            ax1.set_xlabel('Fichiers')
            ax1.set_ylabel('Nombre de cellules')
            ax1.set_title('Nombre Total de Cellules par Fichier', fontweight='bold')
            ax1.set_xticks(range(len(file_names)))
            ax1.set_xticklabels(file_names, rotation=45, ha='right')
            
            # Graphique 2: RÃ©partition text/numeric
            width = 0.35
            x = np.arange(len(file_names))
            ax2.bar(x - width/2, text_cells, width, label='Cellules texte', color='lightcoral', alpha=0.7)
            ax2.bar(x + width/2, numeric_cells, width, label='Cellules numÃ©riques', color='lightgreen', alpha=0.7)
            ax2.set_xlabel('Fichiers')
            ax2.set_ylabel('Nombre de cellules')
            ax2.set_title('RÃ©partition Text/NumÃ©rique par Fichier', fontweight='bold')
            ax2.set_xticks(x)
            ax2.set_xticklabels(file_names, rotation=45, ha='right')
            ax2.legend()
            
            plt.tight_layout()
            plt.savefig(f"{output_dir}/file_statistics.png", dpi=300, bbox_inches='tight')
            plt.close()
        
        print(f"   âœ… Visualisations sauvegardÃ©es dans {output_dir}/")
    
    def _save_results(self, analysis_results: Dict, output_dir: str):
        """Sauvegarde tous les rÃ©sultats"""
        
        # Sauvegarder l'analyse JSON
        with open(f"{output_dir}/analysis_results.json", 'w', encoding='utf-8') as f:
            # Convertir les arrays numpy en listes pour JSON
            results_serializable = self._make_json_serializable(analysis_results)
            json.dump(results_serializable, f, indent=2, ensure_ascii=False)
        
        # Sauvegarder les embeddings
        if self.global_embeddings is not None:
            torch.save(self.global_embeddings, f"{output_dir}/global_embeddings.pt")
        
        # Sauvegarder les embeddings par fichier
        embeddings_by_file_serializable = {}
        for file_path, embeddings in self.embeddings_by_file.items():
            file_name = os.path.basename(file_path)
            torch.save(embeddings, f"{output_dir}/embeddings_{file_name}.pt")
            embeddings_by_file_serializable[file_name] = f"embeddings_{file_name}.pt"
        
        # Sauvegarder le modÃ¨le
        torch.save(self.model.state_dict(), f"{output_dir}/transformer_model.pt")
        
        # Sauvegarder la configuration
        config = {
            'model_config': self.model_config,
            'batch_size': self.batch_size,
            'files_processed': list(self.file_metadata.keys()),
            'embedding_files': embeddings_by_file_serializable,
            'vocabulary_size': len(self.model.string_to_id) if self.model else 0
        }
        
        with open(f"{output_dir}/processing_config.json", 'w') as f:
            json.dump(config, f, indent=2)
        
        # Rapport de synthÃ¨se
        self._generate_summary_report(analysis_results, output_dir)
    
    def _make_json_serializable(self, obj):
        """Convertit les objets numpy/torch en types JSON-sÃ©rialisables"""
        if isinstance(obj, dict):
            return {key: self._make_json_serializable(value) for key, value in obj.items()}
        elif isinstance(obj, list):
            return [self._make_json_serializable(item) for item in obj]
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        elif isinstance(obj, (np.int64, np.int32)):
            return int(obj)
        elif isinstance(obj, (np.float64, np.float32)):
            return float(obj)
        else:
            return obj
    
    def _generate_summary_report(self, analysis_results: Dict, output_dir: str):
        """GÃ©nÃ¨re un rapport de synthÃ¨se"""
        
        report_lines = [
            "ðŸ“Š RAPPORT DE SYNTHÃˆSE - ANALYSE EXCEL EN LOT",
            "=" * 60,
            "",
            f"ðŸ“… Date de traitement: {pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')}",
            f"ðŸ“ Nombre de fichiers traitÃ©s: {len(self.file_metadata)}",
            f"ðŸ“‹ Nombre total de cellules: {len(self.all_cells)}",
            f"ðŸ§  Dimension des embeddings: {self.model_config['embedding_dim']}",
            f"ðŸ“– Taille du vocabulaire: {len(self.model.string_to_id) if self.model else 0}",
            "",
            "ðŸ“ˆ STATISTIQUES PAR FICHIER:",
            "-" * 30
        ]
        
        for file_path, stats in analysis_results.get('file_statistics', {}).items():
            file_name = os.path.basename(file_path)
            report_lines.extend([
                f"",
                f"ðŸ“„ {file_name}:",
                f"  â€¢ Cellules totales: {stats.get('total_cells', 0)}",
                f"  â€¢ Cellules texte: {stats.get('text_cells', 0)}",
                f"  â€¢ Cellules numÃ©riques: {stats.get('numeric_cells', 0)}",
                f"  â€¢ Styles uniques: {stats.get('unique_styles', 0)}",
                f"  â€¢ Norme moyenne embedding: {stats.get('mean_norm', 0):.3f}",
                f"  â€¢ DiversitÃ© embedding: {stats.get('embedding_diversity', 0):.3f}"
            ])
        
        # Clustering
        clusters = analysis_results.get('clusters', {})
        if clusters:
            report_lines.extend([
                "",
                "ðŸŽ¯ CLUSTERING DES FICHIERS:",
                "-" * 25
            ])
            
            for file_name, cluster_id in clusters.get('file_to_cluster', {}).items():
                report_lines.append(f"  â€¢ {file_name}: Cluster {cluster_id}")
        
        # Ã‰crire le rapport
        with open(f"{output_dir}/summary_report.txt", 'w', encoding='utf-8') as f:
            f.write('\n'.join(report_lines))
        
        # Stocker pour les visualisations
        self._last_analysis_results = analysis_results

# === FONCTION UTILITAIRE POUR UTILISATION FACILE ===

def process_excel_batch(excel_files_pattern: str, 
                       output_dir: str = "excel_analysis_results",
                       model_config: Dict = None) -> Dict:
    """
    Fonction utilitaire pour traiter un lot de fichiers Excel
    
    Args:
        excel_files_pattern: Pattern glob pour les fichiers (ex: "data/*.json")
        output_dir: Dossier de sortie
        model_config: Configuration du modÃ¨le (optionnel)
        
    Returns:
        RÃ©sultats de l'analyse
    """
    
    # Trouver les fichiers
    file_paths = glob.glob(excel_files_pattern)
    
    if not file_paths:
        raise ValueError(f"Aucun fichier trouvÃ© avec le pattern: {excel_files_pattern}")
    
    print(f"ðŸ” Fichiers trouvÃ©s: {len(file_paths)}")
    for path in file_paths:
        print(f"  â€¢ {os.path.basename(path)}")
    
    # Traitement
    processor = BatchExcelProcessor(model_config=model_config)
    results = processor.process_excel_files(file_paths, output_dir)
    
    return results

# === EXEMPLE D'UTILISATION ===

def demonstrate_batch_processing():
    """DÃ©montre le traitement en lot avec des fichiers d'exemple"""
    
    print("ðŸŽ¯ DÃ‰MONSTRATION DU TRAITEMENT EN LOT")
    print("=" * 50)
    
    # CrÃ©er des fichiers d'exemple si pas disponibles
    #example_files = create_example_excel_files()
    
    # Configuration du modÃ¨le (plus petite pour la dÃ©mo)
    model_config = {
        'embedding_dim': 64,
        'num_heads': 4,
        'num_layers': 2,
        'vocab_size': 5000,
        'dropout': 0.1
    }
    
    # Traitement en lot
    try:
        results = process_excel_batch(
            excel_files_pattern="*.json",
            output_dir="batch_analysis_demo",
            model_config=model_config
        )
        
        print(f"\nðŸŽ‰ Analyse terminÃ©e avec succÃ¨s !")
        print(f"ðŸ“ RÃ©sultats disponibles dans: batch_analysis_demo/")
        
        return results
        
    except Exception as e:
        print(f"âŒ Erreur: {e}")
        return None

def create_example_excel_files(n_files: int = 5) -> List[str]:
    """CrÃ©e des fichiers Excel JSON d'exemple pour la dÃ©mo"""
    
    example_files = []
    
    for i in range(n_files):
        # Simuler des donnÃ©es Excel diffÃ©rentes
        excel_data = {
            'styles': {
                's0': {'bl': 0},
                's1': {'bl': 1, 'bg': {'rgb': '#4470C4'}},
                's2': {'bd': {'t': {'s': 1}}}
            },
            'sheets': {
                f'sheet_{i}': {
                    'cellData': {}
                }
            }
        }
        
        # GÃ©nÃ©rer des cellules d'exemple
        cell_data = {}
        for row in range(1, 20):  # 20 lignes
            row_data = {}
            for col in range(1, 6):  # 5 colonnes
                
                if row == 1:  # Headers
                    row_data[str(col)] = {
                        'v': f'Header_{col}_File_{i}',
                        't': 1,
                        's': 's1'
                    }
                else:  # DonnÃ©es
                    if col <= 2:  # Colonnes texte
                        row_data[str(col)] = {
                            'v': f'Data_{row}_{col}_F{i}',
                            't': 1,
                            's': 's0'
                        }
                    else:  # Colonnes numÃ©riques
                        row_data[str(col)] = {
                            'v': np.random.randint(1000, 50000) * (i + 1),
                            't': 2,
                            's': 's2'
                        }
            
            cell_data[str(row)] = row_data
        
        excel_data['sheets'][f'sheet_{i}']['cellData'] = cell_data
        
        # Sauvegarder
        filename = f"example_excel_{i}.json"
        with open(filename, 'w', encoding='utf-8') as f:
            json.dump(excel_data, f, indent=2)
        
        example_files.append(filename)
    
    print(f"âœ… {n_files} fichiers d'exemple crÃ©Ã©s")
    return example_files

# === INSTRUCTIONS D'UTILISATION ===

if __name__ == "__main__":
    print("""
ðŸš€ TRAITEMENT EN LOT D'EXCEL - INSTRUCTIONS

1. Pour traiter VOS fichiers Excel JSON:
   
   results = process_excel_batch(
       excel_files_pattern="path/to/your/excel/*.json",
       output_dir="mes_resultats_excel"
   )

2. Pour la dÃ©monstration avec fichiers d'exemple:
   
   results = demonstrate_batch_processing()

3. Structure des rÃ©sultats gÃ©nÃ©rÃ©s:
   ðŸ“ output_dir/
   â”œâ”€â”€ ðŸ“Š analysis_results.json      # Analyses complÃ¨tes
   â”œâ”€â”€ ðŸ§  global_embeddings.pt       # Embeddings globaux
   â”œâ”€â”€ ðŸ“ˆ similarity_matrix.png      # Matrice similaritÃ©
   â”œâ”€â”€ ðŸ—ºï¸ embeddings_tsne.png        # Visualisation t-SNE
   â”œâ”€â”€ ðŸ“Š file_statistics.png        # Stats par fichier
   â”œâ”€â”€ ðŸ“„ summary_report.txt          # Rapport de synthÃ¨se
   â””â”€â”€ âš™ï¸ processing_config.json      # Configuration

ðŸŽ¯ PrÃªt pour analyser une dizaine de fichiers Excel !
""")
    
    # Lancer la dÃ©monstration
    demonstrate_batch_processing()
```

```python
# ============================================================================
# TRAITEMENT EN LOT DE FICHIERS EXCEL JSON - VERSION JUPYTER INTÃ‰GRÃ‰E
# ============================================================================
# Version intÃ©grÃ©e avec analyse de clusters interactive pour Jupyter notebook
# ============================================================================

import os
import json
import glob
from pathlib import Path
from typing import List, Dict, Tuple, Optional
import pandas as pd
import numpy as np
import torch
from concurrent.futures import ThreadPoolExecutor, as_completed
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans, DBSCAN
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from sklearn.neighbors import NearestNeighbors

# Configuration pour Jupyter
%matplotlib inline
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Importation optionnelle de Plotly pour visualisations interactives
try:
    import plotly.express as px
    import plotly.graph_objects as go
    from plotly.subplots import make_subplots
    PLOTLY_AVAILABLE = True
except ImportError:
    PLOTLY_AVAILABLE = False
    print("â„¹ï¸ Plotly non disponible - utilisation de matplotlib uniquement")

import warnings
warnings.filterwarnings('ignore')

# Supposer que les classes prÃ©cÃ©dentes sont disponibles
# from previous_cells import FullCellInfo, ExcelParser, TransformerCellEmbedder, etc.

class JupyterBatchExcelProcessor:
    """
    Version intÃ©grÃ©e du processeur avec analyse de clusters pour Jupyter
    """
    
    def __init__(self, 
                 model_config: Dict = None,
                 batch_size: int = 32,
                 max_workers: int = 4):
        
        self.batch_size = batch_size
        self.max_workers = max_workers
        
        # Configuration par dÃ©faut du modÃ¨le
        self.model_config = model_config or {
            'embedding_dim': 128,
            'num_heads': 8,
            'num_layers': 4,
            'vocab_size': 10000,
            'dropout': 0.1
        }
        
        # ModÃ¨le et donnÃ©es
        self.model = None
        self.all_cells = []
        self.file_metadata = {}
        self.embeddings_by_file = {}
        self.global_embeddings = None
        
        # RÃ©sultats d'analyse
        self.cluster_results = {}
        self.tsne_results = None
        self.pca_results = None
        self.cluster_analysis = None
        
        # Configuration d'affichage
        self.show_plots = True
        
    def process_excel_files_jupyter(self, 
                                   file_paths: List[str],
                                   analyze_clusters: bool = True,
                                   max_clusters: int = 5,
                                   save_results: bool = False,
                                   output_dir: str = "excel_analysis_results") -> Dict:
        """
        Traite un lot de fichiers Excel JSON avec analyse de clusters intÃ©grÃ©e
        
        Args:
            file_paths: Liste des chemins vers les fichiers JSON
            analyze_clusters: Si True, effectue l'analyse de clusters
            max_clusters: Nombre maximum de clusters Ã  tester
            save_results: Si True, sauvegarde les rÃ©sultats sur disque
            output_dir: Dossier de sortie pour les rÃ©sultats
            
        Returns:
            Dict avec rÃ©sultats de l'analyse
        """
        
        print(f"ðŸš€ TRAITEMENT EN LOT DE {len(file_paths)} FICHIERS EXCEL")
        print("=" * 60)
        
        if save_results:
            Path(output_dir).mkdir(exist_ok=True)
        
        # 1. Charger et parser tous les fichiers
        print("\nðŸ“‚ 1. Chargement des fichiers...")
        all_cells_data = self._load_excel_files(file_paths)
        
        # 2. Construire le vocabulaire global
        print("\nðŸ“š 2. Construction du vocabulaire global...")
        self._build_global_model(all_cells_data)
        
        # 3. GÃ©nÃ©rer les embeddings pour tous les fichiers
        print("\nðŸ§  3. GÃ©nÃ©ration des embeddings...")
        self._generate_all_embeddings(all_cells_data)
        
        # 4. Analyses de base
        print("\nðŸ“Š 4. Analyses statistiques de base...")
        basic_analysis = self._perform_basic_analysis()
        
        # 5. Analyse de clusters (nouveau !)
        if analyze_clusters and len(self.embeddings_by_file) > 1:
            print("\nðŸŽ¯ 5. Analyse des clusters...")
            self.cluster_analysis = self._analyze_clusters_comprehensive(max_clusters)
        
        # 6. Visualisations intÃ©grÃ©es
        print("\nðŸ“ˆ 6. GÃ©nÃ©ration des visualisations...")
        self._generate_jupyter_visualizations()
        
        # 7. Sauvegarder si demandÃ©
        if save_results:
            print(f"\nðŸ’¾ 7. Sauvegarde des rÃ©sultats dans {output_dir}...")
            self._save_results(basic_analysis, output_dir)
        
        print(f"\nâœ… Traitement terminÃ© !")
        
        # Retourner tous les rÃ©sultats
        results = {
            'basic_analysis': basic_analysis,
            'cluster_analysis': self.cluster_analysis,
            'embeddings_info': {
                'global_shape': list(self.global_embeddings.shape) if self.global_embeddings is not None else None,
                'files_count': len(self.embeddings_by_file),
                'total_cells': len(self.all_cells)
            }
        }
        
        return results
    
    # === MÃ‰THODES REPRISES DU CODE ORIGINAL ===
    
    def _load_excel_files(self, file_paths: List[str]) -> Dict[str, List]:
        """Charge tous les fichiers Excel en parallÃ¨le (code original)"""
        
        all_cells_data = {}
        
        def load_single_file(file_path: str) -> Tuple[str, List, Dict]:
            try:
                with open(file_path, 'r', encoding='utf-8') as f:
                    excel_data = json.load(f)
                
                # Parser le fichier (suppose que ExcelParser existe)
                cells = ExcelParser.parse_excel_json(excel_data)
                
                # MÃ©tadonnÃ©es du fichier
                metadata = {
                    'file_name': os.path.basename(file_path),
                    'file_path': file_path,
                    'total_cells': len(cells),
                    'text_cells': sum(1 for c in cells if c.cell_type == 1),
                    'numeric_cells': sum(1 for c in cells if c.cell_type == 2),
                    'empty_cells': sum(1 for c in cells if c.cell_type == 0),
                    'formula_cells': sum(1 for c in cells if c.cell_type == 3),
                    'unique_styles': len(set(c.style_id for c in cells)),
                    'sheets': list(excel_data.get('sheets', {}).keys())
                }
                
                return file_path, cells, metadata
                
            except Exception as e:
                print(f"âŒ Erreur lors du chargement de {file_path}: {e}")
                return file_path, [], {}
        
        # Traitement en parallÃ¨le
        with ThreadPoolExecutor(max_workers=self.max_workers) as executor:
            future_to_file = {
                executor.submit(load_single_file, file_path): file_path 
                for file_path in file_paths
            }
            
            for future in tqdm(as_completed(future_to_file), 
                             total=len(file_paths), 
                             desc="Chargement fichiers"):
                file_path, cells, metadata = future.result()
                
                if cells:
                    all_cells_data[file_path] = cells
                    self.file_metadata[file_path] = metadata
                    self.all_cells.extend(cells)
        
        print(f"   âœ… {len(all_cells_data)} fichiers chargÃ©s avec succÃ¨s")
        print(f"   âœ… {len(self.all_cells)} cellules totales extraites")
        
        return all_cells_data
    
    def _build_global_model(self, all_cells_data: Dict[str, List]):
        """Construit le modÃ¨le avec vocabulaire global (code original)"""
        
        print("   ðŸ”¨ Initialisation du TransformerCellEmbedder...")
        # Suppose que device est dÃ©fini globalement
        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        self.model = TransformerCellEmbedder(**self.model_config).to(device)
        
        print("   ðŸ“– Construction du vocabulaire global...")
        self.model.build_vocabulary(self.all_cells)
        
        print(f"   âœ… ModÃ¨le prÃªt avec vocabulaire de {len(self.model.string_to_id)} termes")
    
    def _generate_all_embeddings(self, all_cells_data: Dict[str, List]):
        """GÃ©nÃ¨re les embeddings pour tous les fichiers (code original)"""
        
        self.model.eval()
        all_embeddings = []
        
        with torch.no_grad():
            for file_path, cells in tqdm(all_cells_data.items(), desc="GÃ©nÃ©ration embeddings"):
                
                if not cells:
                    continue
                
                file_embeddings = []
                
                for i in range(0, len(cells), self.batch_size):
                    batch_cells = cells[i:i + self.batch_size]
                    batch_embeddings = self.model(batch_cells)
                    file_embeddings.append(batch_embeddings)
                
                if file_embeddings:
                    file_embeddings_tensor = torch.cat(file_embeddings, dim=0)
                    self.embeddings_by_file[file_path] = file_embeddings_tensor
                    all_embeddings.append(file_embeddings_tensor)
        
        if all_embeddings:
            self.global_embeddings = torch.cat(all_embeddings, dim=0)
        
        print(f"   âœ… Embeddings gÃ©nÃ©rÃ©s: {self.global_embeddings.shape}")
    
    def _perform_basic_analysis(self) -> Dict:
        """Effectue les analyses de base (version simplifiÃ©e)"""
        
        analysis_results = {
            'file_statistics': {},
            'similarity_matrix': None,
            'vocabulary_analysis': {},
            'embedding_statistics': {}
        }
        
        # Statistiques par fichier
        for file_path, metadata in self.file_metadata.items():
            embeddings = self.embeddings_by_file.get(file_path)
            
            if embeddings is not None:
                embedding_stats = {
                    'mean_norm': float(embeddings.norm(dim=1).mean()),
                    'std_norm': float(embeddings.norm(dim=1).std()),
                    'embedding_diversity': float(embeddings.std(dim=0).mean()),
                    'embedding_shape': list(embeddings.shape)
                }
                
                analysis_results['file_statistics'][file_path] = {
                    **metadata,
                    **embedding_stats
                }
        
        # Matrice de similaritÃ©
        file_paths = list(self.embeddings_by_file.keys())
        n_files = len(file_paths)
        similarity_matrix = np.zeros((n_files, n_files))
        
        for i, file1 in enumerate(file_paths):
            for j, file2 in enumerate(file_paths):
                if i <= j:
                    emb1 = self.embeddings_by_file[file1]
                    emb2 = self.embeddings_by_file[file2]
                    
                    similarity = torch.cosine_similarity(
                        emb1.mean(dim=0, keepdim=True),
                        emb2.mean(dim=0, keepdim=True)
                    ).item()
                    
                    similarity_matrix[i, j] = similarity
                    similarity_matrix[j, i] = similarity
        
        analysis_results['similarity_matrix'] = {
            'matrix': similarity_matrix,
            'file_names': [os.path.basename(f) for f in file_paths],
            'file_paths': file_paths
        }
        
        return analysis_results
    
    # === NOUVELLES MÃ‰THODES POUR L'ANALYSE DE CLUSTERS ===
    
    def _analyze_clusters_comprehensive(self, max_clusters: int = 5) -> Dict:
        """
        Analyse complÃ¨te des clusters avec visualisations intÃ©grÃ©es
        """
        
        print("   ðŸ” Analyse complÃ¨te des clusters...")
        
        # 1. PrÃ©paration des donnÃ©es
        file_representations, file_info = self._prepare_file_representations()
        
        # 2. DÃ©termination du nombre optimal de clusters
        optimal_k = self._find_optimal_clusters(file_representations, max_clusters)
        
        # 3. Clustering avec diffÃ©rents algorithmes
        cluster_results = self._perform_clustering(file_representations, file_info, optimal_k)
        
        # 4. RÃ©duction de dimensionnalitÃ©
        self._perform_dimensionality_reduction(file_representations)
        
        # 5. Analyse dÃ©taillÃ©e des clusters
        cluster_analysis = self._analyze_cluster_contents(cluster_results, file_info)
        
        # Stocker les rÃ©sultats
        self.cluster_results = cluster_results
        
        return cluster_analysis
    
    def _prepare_file_representations(self):
        """PrÃ©pare les reprÃ©sentations des fichiers pour le clustering"""
        
        file_representations = []
        file_info = []
        
        for file_path, embeddings in self.embeddings_by_file.items():
            # ReprÃ©sentation moyenne du fichier
            file_rep = embeddings.mean(dim=0).cpu().numpy()
            file_representations.append(file_rep)
            
            # Informations du fichier
            metadata = self.file_metadata.get(file_path, {})
            info = {
                'file_name': os.path.basename(file_path),
                'file_path': file_path,
                'embedding_shape': list(embeddings.shape),
                'avg_norm': float(embeddings.norm(dim=1).mean()),
                'std_norm': float(embeddings.norm(dim=1).std()),
                **metadata
            }
            file_info.append(info)
        
        return np.array(file_representations), file_info
    
    def _find_optimal_clusters(self, representations, max_clusters):
        """Trouve le nombre optimal de clusters avec visualisation"""
        
        n_files = len(representations)
        max_k = min(max_clusters, n_files - 1)
        
        if n_files < 2:
            return 1
        
        inertias = []
        silhouette_scores = []
        k_range = range(2, max_k + 1)
        
        for k in k_range:
            kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
            labels = kmeans.fit_predict(representations)
            
            inertias.append(kmeans.inertia_)
            if len(set(labels)) > 1:
                sil_score = silhouette_score(representations, labels)
                silhouette_scores.append(sil_score)
            else:
                silhouette_scores.append(0)
        
        # MÃ©thode du coude
        if len(inertias) > 1:
            second_derivatives = []
            for i in range(1, len(inertias) - 1):
                second_deriv = inertias[i-1] - 2*inertias[i] + inertias[i+1]
                second_derivatives.append(second_deriv)
            
            if second_derivatives:
                elbow_k = k_range[np.argmax(second_derivatives) + 1]
            else:
                elbow_k = 2
        else:
            elbow_k = 2
        
        # Meilleur silhouette score
        best_sil_k = k_range[np.argmax(silhouette_scores)] if silhouette_scores else 2
        
        # Visualisation des mÃ©triques
        if self.show_plots:
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
            
            # MÃ©thode du coude
            ax1.plot(k_range, inertias, 'bo-', linewidth=2, markersize=8)
            ax1.axvline(x=elbow_k, color='red', linestyle='--', alpha=0.7, label=f'Coude k={elbow_k}')
            ax1.set_xlabel('Nombre de clusters (k)')
            ax1.set_ylabel('Inertie')
            ax1.set_title('MÃ©thode du Coude')
            ax1.legend()
            ax1.grid(True, alpha=0.3)
            
            # Silhouette score
            ax2.plot(k_range, silhouette_scores, 'go-', linewidth=2, markersize=8)
            ax2.axvline(x=best_sil_k, color='red', linestyle='--', alpha=0.7, label=f'Max silhouette k={best_sil_k}')
            ax2.set_xlabel('Nombre de clusters (k)')
            ax2.set_ylabel('Score de Silhouette')
            ax2.set_title('Score de Silhouette')
            ax2.legend()
            ax2.grid(True, alpha=0.3)
            
            plt.tight_layout()
            plt.show()
        
        # Choisir k optimal
        optimal_k = best_sil_k if silhouette_scores else elbow_k
        print(f"      âœ… Nombre optimal de clusters: {optimal_k}")
        
        return optimal_k
    
    def _perform_clustering(self, representations, file_info, optimal_k):
        """Effectue le clustering avec diffÃ©rents algorithmes"""
        
        results = {}
        
        # K-Means
        kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
        kmeans_labels = kmeans.fit_predict(representations)
        
        results['kmeans'] = {
            'labels': kmeans_labels,
            'centers': kmeans.cluster_centers_,
            'inertia': kmeans.inertia_,
            'algorithm': 'K-Means',
            'file_info': file_info.copy(),
            'file_to_cluster': {
                info['file_name']: int(label) 
                for info, label in zip(file_info, kmeans_labels)
            }
        }
        
        # DBSCAN pour comparaison
        if len(representations) > 3:
            neighbors = NearestNeighbors(n_neighbors=2)
            neighbors_fit = neighbors.fit(representations)
            distances, indices = neighbors_fit.kneighbors(representations)
            distances = np.sort(distances[:, 1], axis=0)
            eps = np.percentile(distances, 75)
            
            dbscan = DBSCAN(eps=eps, min_samples=max(2, len(representations)//3))
            dbscan_labels = dbscan.fit_predict(representations)
            
            results['dbscan'] = {
                'labels': dbscan_labels,
                'eps': eps,
                'n_clusters': len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0),
                'algorithm': 'DBSCAN',
                'file_info': file_info.copy(),
                'file_to_cluster': {
                    info['file_name']: int(label) 
                    for info, label in zip(file_info, dbscan_labels)
                }
            }
        
        return results
    
    def _perform_dimensionality_reduction(self, representations):
        """Effectue la rÃ©duction de dimensionnalitÃ©"""
        
        # PCA
        pca = PCA(n_components=2)
        pca_coords = pca.fit_transform(representations)
        
        self.pca_results = {
            'coordinates': pca_coords,
            'explained_variance': pca.explained_variance_ratio_,
            'total_variance': pca.explained_variance_ratio_.sum()
        }
        
        # t-SNE
        perplexity = min(30, len(representations) - 1)
        if perplexity >= 1:
            tsne = TSNE(n_components=2, random_state=42, perplexity=perplexity)
            tsne_coords = tsne.fit_transform(representations)
            
            self.tsne_results = {
                'coordinates': tsne_coords,
                'perplexity': perplexity
            }
    
    def _analyze_cluster_contents(self, cluster_results, file_info):
        """Analyse dÃ©taillÃ©e du contenu des clusters"""
        
        analysis = {}
        
        for method, result in cluster_results.items():
            labels = result['labels']
            
            clusters_analysis = {}
            
            for cluster_id in set(labels):
                if cluster_id == -1:  # Bruit dans DBSCAN
                    continue
                
                mask = labels == cluster_id
                cluster_files = [info for info, is_in_cluster in zip(file_info, mask) if is_in_cluster]
                
                if not cluster_files:
                    continue
                
                # Statistiques du cluster
                cluster_stats = {
                    'n_files': len(cluster_files),
                    'files': [f['file_name'] for f in cluster_files],
                    'total_cells_avg': np.mean([f.get('total_cells', 0) for f in cluster_files]),
                    'text_cells_avg': np.mean([f.get('text_cells', 0) for f in cluster_files]),
                    'numeric_cells_avg': np.mean([f.get('numeric_cells', 0) for f in cluster_files]),
                    'avg_norm_avg': np.mean([f.get('avg_norm', 0) for f in cluster_files]),
                    'std_norm_avg': np.mean([f.get('std_norm', 0) for f in cluster_files]),
                }
                
                # CaractÃ©ristiques dominantes
                cluster_stats['characteristics'] = self._identify_cluster_characteristics(cluster_files)
                
                clusters_analysis[f'cluster_{cluster_id}'] = cluster_stats
            
            analysis[method] = {
                'clusters': clusters_analysis,
                'n_clusters': len(clusters_analysis),
                'algorithm_info': result.get('algorithm', method)
            }
        
        return analysis
    
    def _identify_cluster_characteristics(self, cluster_files):
        """Identifie les caractÃ©ristiques dominantes d'un cluster"""
        
        characteristics = []
        
        avg_total = np.mean([f.get('total_cells', 0) for f in cluster_files])
        avg_text_ratio = np.mean([f.get('text_cells', 0) / max(1, f.get('total_cells', 1)) for f in cluster_files])
        avg_numeric_ratio = np.mean([f.get('numeric_cells', 0) / max(1, f.get('total_cells', 1)) for f in cluster_files])
        
        # Taille des fichiers
        if avg_total > 100:
            characteristics.append("ðŸ“Š Fichiers volumineux")
        elif avg_total < 20:
            characteristics.append("ðŸ“„ Fichiers compacts")
        
        # Type de contenu dominant
        if avg_text_ratio > 0.6:
            characteristics.append("ðŸ“ Majoritairement textuel")
        elif avg_numeric_ratio > 0.6:
            characteristics.append("ðŸ”¢ Majoritairement numÃ©rique")
        else:
            characteristics.append("ðŸ”€ Contenu mixte")
        
        # ComplexitÃ© des embeddings
        avg_norm = np.mean([f.get('avg_norm', 0) for f in cluster_files])
        if avg_norm > 10:
            characteristics.append("ðŸŽ¯ Embeddings complexes")
        elif avg_norm < 5:
            characteristics.append("ðŸŽˆ Embeddings simples")
        
        # Styles
        avg_styles = np.mean([f.get('unique_styles', 0) for f in cluster_files])
        if avg_styles > 5:
            characteristics.append("ðŸŽ¨ Styles variÃ©s")
        elif avg_styles <= 2:
            characteristics.append("ðŸŽ¯ Styles uniformes")
        
        return characteristics
    
    # === VISUALISATIONS POUR JUPYTER ===
    
    def _generate_jupyter_visualizations(self):
        """GÃ©nÃ¨re toutes les visualisations pour Jupyter"""
        
        if not self.show_plots:
            return
        
        # 1. Statistiques de base
        self._plot_basic_statistics()
        
        # 2. Matrice de similaritÃ©
        self._plot_similarity_matrix()
        
        # 3. Visualisations des clusters si disponibles
        if self.cluster_results and (self.pca_results or self.tsne_results):
            self._plot_cluster_visualizations()
            
        # 4. Rapport de clusters
        if self.cluster_analysis:
            self._generate_cluster_report()
    
    def _plot_basic_statistics(self):
        """Graphiques des statistiques de base"""
        
        if not self.file_metadata:
            return
        
        file_names = [os.path.basename(f) for f in self.file_metadata.keys()]
        total_cells = [meta['total_cells'] for meta in self.file_metadata.values()]
        text_cells = [meta['text_cells'] for meta in self.file_metadata.values()]
        numeric_cells = [meta['numeric_cells'] for meta in self.file_metadata.values()]
        
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))
        
        # Nombre total de cellules
        ax1.bar(range(len(file_names)), total_cells, color='skyblue', alpha=0.7)
        ax1.set_xlabel('Fichiers')
        ax1.set_ylabel('Nombre de cellules')
        ax1.set_title('Nombre Total de Cellules par Fichier', fontweight='bold')
        ax1.set_xticks(range(len(file_names)))
        ax1.set_xticklabels(file_names, rotation=45, ha='right')
        
        # RÃ©partition text/numeric
        width = 0.35
        x = np.arange(len(file_names))
        ax2.bar(x - width/2, text_cells, width, label='Cellules texte', color='lightcoral', alpha=0.7)
        ax2.bar(x + width/2, numeric_cells, width, label='Cellules numÃ©riques', color='lightgreen', alpha=0.7)
        ax2.set_xlabel('Fichiers')
        ax2.set_ylabel('Nombre de cellules')
        ax2.set_title('RÃ©partition Text/NumÃ©rique par Fichier', fontweight='bold')
        ax2.set_xticks(x)
        ax2.set_xticklabels(file_names, rotation=45, ha='right')
        ax2.legend()
        
        plt.tight_layout()
        plt.show()
    
    def _plot_similarity_matrix(self):
        """Affiche la matrice de similaritÃ©"""
        
        # Recalculer la matrice de similaritÃ© pour l'affichage
        file_paths = list(self.embeddings_by_file.keys())
        n_files = len(file_paths)
        
        if n_files < 2:
            return
        
        similarity_matrix = np.zeros((n_files, n_files))
        
        for i, file1 in enumerate(file_paths):
            for j, file2 in enumerate(file_paths):
                emb1 = self.embeddings_by_file[file1]
                emb2 = self.embeddings_by_file[file2]
                
                similarity = torch.cosine_similarity(
                    emb1.mean(dim=0, keepdim=True),
                    emb2.mean(dim=0, keepdim=True)
                ).item()
                
                similarity_matrix[i, j] = similarity
        
        file_names = [os.path.basename(f) for f in file_paths]
        
        fig, ax = plt.subplots(figsize=(12, 10))
        sns.heatmap(similarity_matrix, 
                   xticklabels=file_names,
                   yticklabels=file_names,
                   annot=True, 
                   cmap='viridis',
                   center=0.5,
                   ax=ax)
        
        ax.set_title('Matrice de SimilaritÃ© entre Fichiers Excel', fontsize=14, fontweight='bold')
        plt.xticks(rotation=45, ha='right')
        plt.yticks(rotation=0)
        plt.tight_layout()
        plt.show()
    
    def _plot_cluster_visualizations(self):
        """Visualise les clusters en PCA et t-SNE"""
        
        method = 'kmeans'  # Utiliser K-means par dÃ©faut
        if method not in self.cluster_results:
            method = list(self.cluster_results.keys())[0]
        
        result = self.cluster_results[method]
        labels = result['labels']
        file_info = result['file_info']
        file_names = [info['file_name'] for info in file_info]
        
        n_plots = 1 + (1 if self.tsne_results else 0)
        fig, axes = plt.subplots(1, n_plots, figsize=(8*n_plots, 6))
        
        if n_plots == 1:
            axes = [axes]
        
        colors = plt.cm.Set3(np.linspace(0, 1, len(set(labels))))
        
        plot_idx = 0
        
        # PCA
        if self.pca_results:
            ax = axes[plot_idx]
            coords = self.pca_results['coordinates']
            
            for label in set(labels):
                mask = labels == label
                color = colors[label % len(colors)]
                ax.scatter(coords[mask, 0], coords[mask, 1], 
                         c=[color], label=f'Cluster {label}', 
                         s=100, alpha=0.7, edgecolors='black')
            
            # Annoter les points
            for i, name in enumerate(file_names):
                ax.annotate(name[:15], (coords[i, 0], coords[i, 1]), 
                          xytext=(5, 5), textcoords='offset points',
                          fontsize=8, alpha=0.8)
            
            ax.set_xlabel(f'PC1 ({self.pca_results["explained_variance"][0]:.1%})')
            ax.set_ylabel(f'PC2 ({self.pca_results["explained_variance"][1]:.1%})')
            ax.set_title(f'Clusters PCA\n(Variance expliquÃ©e: {self.pca_results["total_variance"]:.1%})')
            ax.legend()
            ax.grid(True, alpha=0.3)
            
            plot_idx += 1
        
        # t-SNE
        if self.tsne_results:
            ax = axes[plot_idx]
            coords = self.tsne_results['coordinates']
            
            for label in set(labels):
                mask = labels == label
                color = colors[label % len(colors)]
                ax.scatter(coords[mask, 0], coords[mask, 1], 
                         c=[color], label=f'Cluster {label}', 
                         s=100, alpha=0.7, edgecolors='black')
            
            # Annoter les points
            for i, name in enumerate(file_names):
                ax.annotate(name[:15], (coords[i, 0], coords[i, 1]), 
                          xytext=(5, 5), textcoords='offset points',
                          fontsize=8, alpha=0.8)
            
            ax.set_xlabel('t-SNE 1')
            ax.set_ylabel('t-SNE 2')
            ax.set_title('Clusters t-SNE')
            ax.legend()
            ax.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
    
    def _generate_cluster_report(self):
        """GÃ©nÃ¨re un rapport dÃ©taillÃ© des clusters"""
        
        if not self.cluster_analysis:
            return
        
        print("\n" + "="*60)
        print("ðŸ“Š RAPPORT DÃ‰TAILLÃ‰ DES CLUSTERS")
        print("="*60)
        
        for method, analysis in self.cluster_analysis.items():
            print(f"\nðŸ” MÃ‰THODE: {analysis['algorithm_info'].upper()}")
            print("-" * 40)
            print(f"Nombre de clusters: {analysis['n_clusters']}")
            
            for cluster_name, cluster_info in analysis['clusters'].items():
                print(f"\nðŸ“‚ {cluster_name.upper()}:")
                print(f"   â€¢ Nombre de fichiers: {cluster_info['n_files']}")
                print(f"   â€¢ Fichiers: {', '.join(cluster_info['files'])}")
                print(f"   â€¢ Cellules moyennes: {cluster_info['total_cells_avg']:.1f}")
                print(f"   â€¢ Ratio texte: {cluster_info['text_cells_avg']/max(1, cluster_info['total_cells_avg']):.1%}")
                print(f"   â€¢ Ratio numÃ©rique: {cluster_info['numeric_cells_avg']/max(1, cluster_info['total_cells_avg']):.1%}")
                print(f"   â€¢ Norme embedding moyenne: {cluster_info['avg_norm_avg']:.3f}")
                
                print("   ðŸ·ï¸ CaractÃ©ristiques:")
                for char in cluster_info['characteristics']:
                    print(f"      {char}")
        
        # Recommandations
        n_clusters = len(list(self.cluster_analysis.values())[0]['clusters'])
        print(f"\nðŸ’¡ RECOMMANDATIONS:")
        print("-" * 20)
        
        if n_clusters <= 2:
            print("â€¢ Les fichiers se regroupent en peu de catÃ©gories distinctes")
            print("â€¢ VÃ©rifiez si vous avez des types de documents similaires")
        elif n_clusters >= 4:
            print("â€¢ Forte diversitÃ© dans vos fichiers Excel")
            print("â€¢ Chaque cluster reprÃ©sente probablement un usage spÃ©cifique")
        
        print("â€¢ Utilisez ces clusters pour organiser vos fichiers")
        print("â€¢ Les caractÃ©ristiques peuvent guider le traitement automatique")
    
    def _save_results(self, analysis_results: Dict, output_dir: str):
        """Sauvegarde les rÃ©sultats (version simplifiÃ©e)"""
        
        # Sauvegarder l'analyse JSON
        results_to_save = {
            'basic_analysis': self._make_json_serializable(analysis_results),
            'cluster_analysis': self._make_json_serializable(self.cluster_analysis) if self.cluster_analysis else None
        }
        
        with open(f"{output_dir}/analysis_results.json", 'w', encoding='utf-8') as f:
            json.dump(results_to_save, f, indent=2, ensure_ascii=False)
        
        # Sauvegarder les embeddings
        if self.global_embeddings is not None:
            torch.save(self.global_embeddings, f"{output_dir}/global_embeddings.pt")
        
        # Sauvegarder le modÃ¨le
        if self.model is not None:
            torch.save(self.model.state_dict(), f"{output_dir}/transformer_model.pt")
    
    def _make_json_serializable(self, obj):
        """Convertit les objets numpy/torch en types JSON-sÃ©rialisables"""
        if isinstance(obj, dict):
            return {key: self._make_json_serializable(value) for key, value in obj.items()}
        elif isinstance(obj, list):
            return [self._make_json_serializable(item) for item in obj]
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        elif isinstance(obj, (np.int64, np.int32)):
            return int(obj)
        elif isinstance(obj, (np.float64, np.float32)):
            return float(obj)
        else:
            return obj

# ============================================================================
# FONCTIONS UTILITAIRES POUR JUPYTER
# ============================================================================

def process_excel_batch_jupyter(excel_files_pattern: str, 
                               analyze_clusters: bool = True,
                               max_clusters: int = 5,
                               save_results: bool = False,
                               output_dir: str = "excel_analysis_results",
                               model_config: Dict = None) -> Tuple[JupyterBatchExcelProcessor, Dict]:
    """
    Fonction principale pour traiter un lot de fichiers Excel dans Jupyter
    
    Args:
        excel_files_pattern: Pattern glob pour les fichiers (ex: "data/*.json")
        analyze_clusters: Si True, effectue l'analyse de clusters
        max_clusters: Nombre maximum de clusters Ã  tester
        save_results: Si True, sauvegarde sur disque
        output_dir: Dossier de sortie
        model_config: Configuration du modÃ¨le (optionnel)
        
    Returns:
        (processor, results): Instance du processeur et rÃ©sultats
    """
    
    # Trouver les fichiers
    file_paths = glob.glob(excel_files_pattern)
    
    if not file_paths:
        raise ValueError(f"Aucun fichier trouvÃ© avec le pattern: {excel_files_pattern}")
    
    print(f"ðŸ” Fichiers trouvÃ©s: {len(file_paths)}")
    for path in file_paths[:5]:  # Afficher max 5 fichiers
        print(f"  â€¢ {os.path.basename(path)}")
    if len(file_paths) > 5:
        print(f"  â€¢ ... et {len(file_paths) - 5} autres")
    
    # Traitement
    processor = JupyterBatchExcelProcessor(model_config=model_config)
    results = processor.process_excel_files_jupyter(
        file_paths=file_paths,
        analyze_clusters=analyze_clusters,
        max_clusters=max_clusters,
        save_results=save_results,
        output_dir=output_dir
    )
    
    return processor, results

def quick_cluster_analysis(processor: JupyterBatchExcelProcessor):
    """
    Analyse rapide des clusters pour un processeur existant
    """
    
    if not processor.cluster_results:
        print("âŒ Aucun rÃ©sultat de clustering disponible. Lancez d'abord l'analyse complÃ¨te.")
        return
    
    print("ðŸŽ¯ ANALYSE RAPIDE DES CLUSTERS")
    print("=" * 40)
    
    # Afficher les assignations de clusters
    for method, result in processor.cluster_results.items():
        print(f"\nðŸ“Š {method.upper()}:")
        for file_name, cluster_id in result['file_to_cluster'].items():
            print(f"  â€¢ {file_name} â†’ Cluster {cluster_id}")
    
    # Reafficher les visualisations
    processor._plot_cluster_visualizations()

# ============================================================================
# EXEMPLE D'UTILISATION COMPLÃˆTE POUR JUPYTER
# ============================================================================

def demo_jupyter_excel_analysis():
    """
    DÃ©monstration complÃ¨te pour Jupyter notebook
    """
    # 1. Traitement complet avec analyse de clusters:
    processor, results = process_excel_batch_jupyter(
        excel_files_pattern="*.json",
        analyze_clusters=True,
        max_clusters=5,
        save_results=False
    )
    
    # 2. AccÃ¨s aux rÃ©sultats:
    print("ðŸ“Š Clusters K-means:")
    print(results['cluster_analysis']['kmeans']['clusters'])
    
    # 3. Analyse rapide supplÃ©mentaire:
    quick_cluster_analysis(processor)
    
    # 4. AccÃ¨s aux donnÃ©es brutes:
    print("Embeddings shape:", processor.global_embeddings.shape)
    print("Files processed:", len(processor.embeddings_by_file))


if __name__ == "__main__":
    demo_jupyter_excel_analysis()
```
